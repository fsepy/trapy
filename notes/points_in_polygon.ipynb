```python
def test_shapely(points, polygon):
    from shapely.geometry import Point
    from shapely.geometry.polygon import Polygon

    polygon = Polygon(polygon)

    points_in_poly_result = list()
    for point in points:
        point = Point(point)
        points_in_poly_result.append(polygon.contains(point))
    
    print(points_in_poly_result)

test_shapely(points, polygon)
```
